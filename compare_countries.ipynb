```python
from scipy.stats import f_oneway

# Perform One-Way ANOVA
result = f_oneway(benin_df['GHI'], sierra_leone_df['GHI'], togo_df['GHI'])

# Extract the F-statistic and p-value
f_statistic = result.statistic
p_value = result.pvalue

# Markdown report
markdown_report = f"""
### Results of One-Way ANOVA

- **F-statistic**: {f_statistic}
- **p-value**: {p_value}

#### Interpretation
- {'Reject the null hypothesis' if p_value < 0.05 else 'Fail to reject the null hypothesis'}
"""

print(markdown_report)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the datasets
benin_df = pd.read_csv(r'C:\Users\hp\Music\data\benin-malanville.csv', encoding='latin-1')
sierra_leone_df = pd.read_csv(r'C:\Users\hp\Music\data\sierraleone-bumbuna.csv', encoding='latin-1')
togo_df = pd.read_csv(r'C:\Users\hp\Music\data\togo-dapaong_qc.csv', encoding='latin-1')

# Calculate Mean GHI for each country
mean_values = {
    'Benin': benin_df['GHI'].mean(),
    'Sierra Leone': sierra_leone_df['GHI'].mean(),
    'Togo': togo_df['GHI'].mean()
}

# Create a DataFrame
mean_ghi_df = pd.DataFrame({'Mean GHI': mean_values})

# Plot the bar chart
mean_ghi_df.plot(kind='bar', legend=False)
plt.title('Mean Global Horizontal Irradiance (GHI) by Country')
plt.ylabel('Mean GHI (W/m²)')
plt.xticks(rotation=45)
plt.show()
```
